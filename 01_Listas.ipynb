## Una **lista** es como una caja ordenada de elementos:

📦 = [elemento1, elemento2, elemento3]

* Tiene orden
* Es modificable (mutable)
* Puede contener cualquier tipo de dato

## 📘 Conceptos formales

| Concepto    | Descripción                         |
| ----------- | ----------------------------------- |
| Mutable     | Se puede modificar después de crear |
| Indexación  | Acceso por posición (`lista[0]`)    |
| Heterogénea | Puede contener distintos tipos      |
| Dinámica    | Puede crecer o reducirse            |

````python
# Crear una lista
numeros = [1, 2, 3, 4, 5, 7, 8, 9, 10, 11, 12, 13, 14, 15]
````

````python
print(numeros)
````

````python
numeros
````

````python
# Lista con diferentes tipos
mixta = [10, "hola", 3.14, True]
````

````python
mixta
````

````python
# Acceder a elementos
print(numeros[0])   # 1
print(numeros[2])   # 3
````

````python
# Acceder desde atrás
print(numeros[-1])  # 5
````

### ✅ Modificar listas

````python
numeros[1] = 20
print(numeros)
````

### ✅ Métodos clave

````python
# Agregar elemento
numeros.append(6)
````

````python
# Insertar en posición específica
numeros.insert(10, 99)
````

````python
# Eliminar elemento
numeros.remove(3)
````

````python
# Eliminar último elemento
numeros.pop()
````

### ✅ Imprime cambios

````python
numeros
````

### ✅ Recorrer listas

````python
for n in numeros:
    print(n)
````

### ✅ List comprehension (MUY importante en Data Science)

````python
cuadrados = [x**2 for x in range(5)]
print(cuadrados)
````
